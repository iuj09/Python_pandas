### 결측치(Missing Value)

 - NA : Not Available(유효하지 않은)
 - NaN : Not a Number(숫자가 아닌)
 - Null : 아무것도 존재하지 않음을 의미

```python
df = pd.read_csv("소상공인시장진흥공단_상가업소정보_의료기관_201909.csv", low_memory=False)

df.shape
```

```python
# df.info() : 데이터 프레임 정보 확인
df.info()
```

```python
# df.isnull() 데이터 프레임에서 결측치를 True, 값이 있으면 False를 반환
df.isnull()
```

```python
null_count = df.isnull().sum()
null_count
```

Series.reset_index() : Series 형태의 자료를 DataFrame 자료로 변형

```python
# .reset_index() --> serise를 data frame으로 변환
df_null_count = null_count.reset_index()
df_null_count.head()
```

```python
# DataFrame.columns = ["바꿀 컬럼명", ..., (반드시 컬럼 갯수만큼!) : DataFrame의 컬럼명을 바꿈
df_null_count.columns =['컬럼명', '결측지수']
df_null_count.head()
```

DataFrame.sort_values( by="컬럼명")

```python
#df에서 결측치가 가장 많은 10개를 df형식으로 가져옴
df_null_count_top10 = df_null_count.sort_values(by="결측지수", ascending=False).head(10)
df_null_count_top10
```

**DataFrame.loc(행,열), 여기서 행과 열은 list나 Series 형태로 입력 가능**

ex) DataFrame.loc[Series(True/False), "컬럼명"]

```python
# medical에는 df에서  "상권업종대분류명"이 "의료"인 것만 True로 표시
# 나머지는 False로 되어 있는 시리즈
medical = df["상권업종대분류명"] == "의료"

# df.loc[Series(True/False),"원하는 컬럼의 이름"]
# "상권업종대분류명"이 "의료"인 자료에서 "상권업종중분류명" 컬럼만 가지는 시리즈 리턴
df.loc[medical, "상권업종중분류명"]
```

## 인덱싱을 이용한 데이터 전처리

```python
df_seoul_hospital = df[(df["상권업종소분류명"] == "종합병원") & (df["시도명"] == "서울특별시")].copy()
df_seoul_hospital.head(2)
```

```python
# '서울특별시'에 위치한 '종합병원'의 상호명 데이터 중, "종합병원"을 포함하는 데이터 인덱싱하기
df_seoul_hospital.loc[~df_seoul_hospital["상호명"].str.contains("종합병원"), "상호명"].unique()
```

```python
# 꽃배달, 의료기, 장례식장, 상담소, 어린이집을 포함한 데이터 drop_row에 저장
drop_rowl = df_seoul_hospital[df_seoul_hospital["상호명"].str.contains("꽃배달|의료기|장례식장|상담소|어린이집")].index
drop_rowl = drop_rowl.tolist()
drop_rowl
```

```python
# '의원'으로 끝나는 데이터 인덱싱
drop_row2 = df_seoul_hospital[df_seoul_hospital["상호명"].str.endswith("의원")].index
drop_row2 = drop_row2.tolist()
drop_row2
```

```python
drop_row = drop_rowl + drop_row2
```

```python
print(df_seoul_hospital.shape)
df_seoul_hospital = df_seoul_hospital.drop(drop_row, axis=0)
print(df_seoul_hospital.shape)
```

### 데이터 전처리

```python
# 분양가격 열(column)의 데이터 타입을 int로 바꿈
df['분양가격'].astype(int)
```

**df.loc[] 함수**

 - df.loc[Series, Column]의 형태로 자주 쓰이며, 특정 조건을 만족하는 데이터를 추출하거나 변경하고 싶을 때 사용 함

```python
# '분양가격' 열의 데이터 중 '  '(빈칸 두 개)로 이루어진 행만 출력
df.loc[df['분양가격'] == '  ']
```

```python
# '분양가격'이 '  '인 것을 0으로 일괄 변경
df.loc[df['분양가격'] == '  ', '분양가격'] = '0'
```

```python
df['분양가격'].astype(int)
```

```python
# NaN값을 fillna() 함수를 이용해 0으로 채워 줌
df['분양가격'] = df['분양가격'].fillna('0')
```

```python
df['분양가격'].astype(int)
```

```python
df.loc[df['분양가격'] == '6,657']
```

```python
# replace() 함수를 통해 ','를 ''(따옴표 사이에 아무것도 없음)로 바꾸어 줌으로써 제거

df['분양가격'] = df['분양가격'].str.replace(',', '')
```

```python
df['분양가격'].astype(int)
```

```python
# replace() 함수를 이용해 '-'를 '0'으로 바꾸어 줌
df['분양가격'] = df['분양가격'].str.replace('-', '0')
```
